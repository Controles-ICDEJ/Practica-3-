```python
import os
import pandas as pd
```

```python
# Ruta a la carpeta de los archivos
append_data_folder = os.path.join('extracted_files', 'appendData')
```

```python
# Diccionario para el data frame a crear 
dataframes = {}
```

```python
# Iteraramos sobre los archivos .xlsx en la carpeta 'appendData'
for filename in os.listdir(append_data_folder):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(append_data_folder, filename)
        # Leer el archivo Excel
        df = pd.read_excel(file_path, engine='openpyxl')
        print(f'Archivo cargado: {filename}')
        
        # Además, guardamos el DataFrame en el diccionario
        dataframes[filename] = df
```

```python
# Asignamos a cada DataFrame a una variable individual para facilitar el acceso
df1 = dataframes.get('level1.xlsx')
df2 = dataframes.get('level2.xlsx')
df3 = dataframes.get('level3.xlsx')
df4 = dataframes.get('level4.xlsx')
```

```python
# Cargar el archivo CSV especificando el delimitador correcto (;)
ruta_archivo = 'fallecidosCOVID.csv'
data = pd.read_csv(ruta_archivo, delimiter=';')
```

```python
# Renombrar las columnas para mayor claridad
data.columns = [
    'FechaCorte', 'FechaFallecimiento', 'EdadDeclarada', 
    'Sexo', 'ClasificacionDef', 'D', 
    'P', 'D', 'Ub', 'Uu'
]
```

```python
# Convertir la columna 'FECHA_FALLECIMIENTO' a formato de fecha
data['FechaFallecimiento'] = pd.to_datetime(data['FechaFallecimiento'], format='%Y%m%d')
```

```python
# Crear una nueva columna para el año y el mes
data['AñoMes'] = data['FechaFallecimiento'].dt.to_period('M')
```

```python
# Agrupar por 'AÑO_MES' y contar el número de fallecidos
reporte_mensual = data.groupby('AñoMes').size().reset_index(name='CantidadFallecidos')
```

```python
reporte_mensual
```

```python
##Parte 4 
```

```python
reporte_tipo_muerte = data.pivot_table(
    index='AñoMes', 
    columns='ClasificacionDef', 
    aggfunc='size', 
    fill_value=0
).reset_index()
```
